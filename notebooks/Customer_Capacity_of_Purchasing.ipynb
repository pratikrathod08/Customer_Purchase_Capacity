## About the dataset ( BankChurened dataset)

1. CLIENTNUM: It likely refers to a unique identifier or a client number assigned to each customer in the dataset.
2. Attrition_Flag: This attribute may indicate whether a customer has churned or left the company (e.g., a binary flag for customer retention).
3. Customer_Age: It represents the age of the customer.
4. Gender: It specifies the gender of the customer (e.g., male or female).
5. Dependent_count: This attribute indicates the number of dependents the customer has.
6. Education_Level: It denotes the educational level or qualification of the customer.
7. Marital_Status: It represents the marital status of the customer (e.g., married, single, divorced).
8. Income_Category: This attribute categorizes the income level of the customer (e.g., low, medium, high).
9. Card_Category: It specifies the category of the credit card held by the customer (e.g., gold, platinum, silver).
10. Months_on_book: It represents the number of months the customer has been a cardholder.
11. Total_Relationship_Count: This attribute indicates the total number of products or services the customer has with the bank.
12. Months_Inactive_12_mon: It denotes the number of months the customer has been inactive in the last 12 months.
13. Contacts_Count_12_mon: This attribute represents the number of contacts the customer had with the bank in the last 12 months.
14. Credit_Limit: It indicates the credit limit assigned to the customer.
15. Total_Revolving_Bal: This attribute represents the total balance on the revolving credit accounts.
16. Avg_Open_To_Buy: It denotes the average amount available for the customer to spend on their credit accounts.
17. Total_Amt_Chng_Q4_Q1: This attribute represents the percentage change in the total transaction amount between the fourth quarter and the first quarter.
18. Total_Trans_Amt: It indicates the total transaction amount in the last 12 months.
19. Total_Trans_Ct: This attribute represents the total number of transactions in the last 12 months.
20. Total_Ct_Chng_Q4_Q1: It denotes the percentage change in the total transaction count between the fourth quarter and the first quarter.
21. Avg_Utilization_Ratio: This attribute represents the average utilization ratio of the customer's credit limit.

## 8 Outlier Treatment with IQR

```python
data['Months_on_book']
q1=data['Months_on_book'].quantile(0.25)
q3=data['Months_on_book'].quantile(0.75)
iqr=q3-q1
lower_month_on_book = q1 -(iqr*1.5)
upper_month_on_book = q3 +(iqr*1.5)
print(lower_month_on_book)
print(upper_month_on_book)
```

```python
data['Months_on_book']=data['Months_on_book'].apply(lambda x : data['Months_on_book'].mean()  if x <= lower_month_on_book or x>= upper_month_on_book else x )
```

```python
sns.boxplot(data['Months_on_book'])
```

```python
def outlier_remover(data,column_name):
    q1=data[column_name].quantile(0.25)
    q3=data[column_name].quantile(0.75)
    iqr=q3-q1
    lower = q1 -(iqr*1.5)
    upper = q3 +(iqr*1.5)
    data[column_name]=data[column_name].apply(lambda x : data[column_name].mean()  if x <= lower or x>= upper else x )
```

```python
outlier_remover(data,"Customer_Age")
sns.boxplot(data["Customer_Age"])
```

```python
outlier_remover(data,"Months_on_book")
sns.boxplot(data["Months_on_book"])
```

```python
num.columns
```

```python
outlier_remover(data,'Total_Relationship_Count')
outlier_remover(data,'Months_Inactive_Count')
outlier_remover(data,'Contacts_Count')
outlier_remover(data,'Credit_Limit')
outlier_remover(data,'Total_Revolving_Bal')
outlier_remover(data,'Avg_Open_To_Buy')
outlier_remover(data,'Total_Trans_Amt')
outlier_remover(data,'Total_Trans_Ct')
outlier_remover(data,'Avg_Utilization_Ratio')
```
